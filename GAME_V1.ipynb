```python
# importando libs 

import random
from os import system, name
```

```python
# função para limpar a tela a cada execução

def limparTela():
    
    # windows 
    if name == 'nt':
        _= system('cls')
    else:
        # mac ou linux
        _= system('clear')
```

```python
# função do jogo

def game():
    
    limparTela()
    
    print('\nBem-vinda(o) ao jogo da forca!')
    print('Advinhe a palavra abaixo: \n')
    
    # lista de palavras para o jogo
    palavras = ['banana','abacate','uva','morango','laranja', 'abacaxi']
    
    # escolhe randomicamente uma palavra 
    palavra = random.choice(palavras)
    
    # list comprehension
    letras_descobertas = ['_' for letra in palavra]
    
    # número de chances (cuidado com o tamanho das palavras)
    chances = 6
    
    # lista para as letras erradas
    letras_erradas = []
    
    # loop enquanto o número de chances for maior do que zero
    while chances > 0:
        
        print(' '.join(letras_descobertas))
        print('\nChances restantes: ', chances)
        print('Letras erradas: ', ' '.join(letras_erradas))
        
        # tentativa
        tentativa = input('\nDigite uma letra: ').lower()
        
        #condicional
        if tentativa in palavra:
            index = 0
            
            for letra in palavra:
                if tentativa == letra:
                    letras_descobertas[index] = letra
                    
                index += 1
        else:
            chances -= 1
            letras_erradas.append(tentativa)
        
        # condicional
        if '_' not in letras_descobertas:
            print('\nVocê venceu, a palavra era: ', palavra)
            break
    
            
    # condicional
    if '_' in letras_descobertas:
        print('\nVocê perdeu, a palavra era: ', palavra)

# bloco main > indica que é um programa python
if __name__ == '__main__':
    game()
    print('\nParabéns. Você está aprendendo programação em Python com a DSA. :)\n')
```
