```python
# importando libs 

import random
from os import system, name
```

```python
# função para limpar a tela a cada execução

def limparTela():
    
    # windows 
    if name == 'nt':
        _= system('cls')
    else:
        # mac ou linux
        _= system('clear')
        
# função que desenha a forca na tela
def display_hangman(chances):
    
    #lista de estágios da forca
    # Lista de estágios da forca
    stages = [  # estágio 6 (final)
                """
                   --------
                   |      |
                   |      O
                   |     \\|/
                   |      |
                   |     / \\
                   -
                """,
                # estágio 5
                """
                   --------
                   |      |
                   |      O
                   |     \\|/
                   |      |
                   |     / 
                   -
                """,
                # estágio 4
                """
                   --------
                   |      |
                   |      O
                   |     \\|/
                   |      |
                   |      
                   -
                """,
                # estágio 3
                """
                   --------
                   |      |
                   |      O
                   |     \\|
                   |      |
                   |     
                   -
                """,
                # estágio 2
                """
                   --------
                   |      |
                   |      O
                   |      |
                   |      |
                   |     
                   -
                """,
                # estágio 1
                """
                   --------
                   |      |
                   |      O
                   |    
                   |      
                   |     
                   -
                """,
                # estágio 0
                """
                   --------
                   |      |
                   |      
                   |    
                   |      
                   |     
                   -
                """
    ]
    return stages[chances]
        
```

```python
# função do jogo

def game():
    
    limparTela()
    
    print('\nBem-vinda(o) ao jogo da forca!')
    print('Advinhe a palavra abaixo: \n')
    
    # lista de palavras para o jogo
    palavras = ['banana','abacate','uva','morango','laranja', 'abacaxi']
    
    # escolhe randomicamente uma palavra 
    palavra = random.choice(palavras)
    
    # lista de letras da palavra
    lista_letras_palavras = [letra for letra in palavra]
    
    # cria o tabuleiro com o caracter '_' multiplicado pelo comprimento da palavra
    tabuleiro = ['_'] * len(palavra)
    
    # número de chances (cuidado com o tamanho das palavras)
    chances = 6
    
    letras_tentativas = []
    
    # loop enquanto o número de chances for maior do que zero
    while chances > 0:
        
        print(display_hangman(chances))
        print('Palavra: ', tabuleiro)
        print('\n')

        # tentativa
        tentativa = input('\nDigite uma letra: ').lower()
        
        #condicional
        if tentativa in letras_tentativas:
            print('Você já tentou essa letra. Escolha outra!')
            continue
            
        # lista de tentativas (letra)
        letras_tentativas.append(tentativa)
            
        # condicional
        if tentativa in lista_letras_palavras:
            
            print('Você acertou a letra!')
            
            #loop
            for indice in range(len(lista_letras_palavras)):
                # condicional
                if lista_letras_palavras[indice] == tentativa:
                    tabuleiro[indice] = tentativa
            
            # condicional
            if '_' not in tabuleiro:
                print('\nVocê venceu, a palavra era: ', palavra)
                break
        else:
            print('Ops, essa letra não está na palavra!')
            chances -= 1
            
      # condicional
    if '_' in lista_letras_palavras:
            print('\nVocê perdeu, a palavra era: ', palavra)
            

# bloco main > indica que é um programa python
if __name__ == '__main__':
    game()
    print('\nParabéns. Você está aprendendo programação em Python com a DSA. :)\n')
```
